```python
x = np.arange(1,10)
x
y = x**2
y
plt.plot(x,y,"r--")
plt.title("Quadratics")
plt.xlabel("x")
plt.ylabel("x^2")
plt.grid()
plt.show()
```

```python
x = np.arange(1,10)
x
y = x**2
y
plt.plot(x,y,"r--")
plt.title("Quadratics")
plt.xlabel("x")
plt.ylabel("x^2")
plt.grid()
plt.xlim(-10,100)
plt.ylim(-1,10)

plt.show()
```

```python
x = np.arange(1,10)
x
y = x**2
y
plt.title("Quadratics")
plt.xlabel("x")
plt.ylabel("x^2")
plt.grid()
plt.text(0,2,"hellooo")

plt.plot(x,y,"r--",label="square",linewidth = 4)
plt.legend()
plt.show()
```

```python
x = np.arange(1,101)
y = np.log(x)
plt.plot(x,y)
plt.show()
```

```python
import math
x = np.arange(1,101)
#y = np.log(x)

y= [math.log(v) for v in x]
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(1,101)
#y = np.log(x)
y = list(map(math.log,x))  #No brackets
plt.plot(x,y)
plt.show()
```
